### 2. Examine the State and Action Spaces

The simulation contains a single agent that navigates a large environment.  At each time step, it has four actions at its disposal:
- `0` - walk forward 
- `1` - walk backward
- `2` - turn left
- `3` - turn right

The state space has `37` dimensions and contains the agent's velocity, along with ray-based perception of objects around agent's forward direction.  A reward of `+1` is provided for collecting a yellow banana, and a reward of `-1` is provided for collecting a blue banana. 

Run the code cell below to print some information about the environment.

````python
# reset the environment
env_info = env.reset(train_mode=True)[brain_name]

# number of agents in the environment
print('Number of agents:', len(env_info.agents))

# number of actions
action_size = brain.vector_action_space_size
print('Number of actions:', action_size)

# examine the state space 
state = env_info.vector_observations[0]
print('States look like:', state)
state_size = len(state)
print('States have length:', state_size)
````

### 3. Take Random Actions in the Environment

In the next code cell, you will learn how to use the Python API to control the agent and receive feedback from the environment.

Once this cell is executed, you will watch the agent's performance, if it selects an action (uniformly) at random with each time step.  A window should pop up that allows you to observe the agent, as it moves through the environment.  

Of course, as part of the project, you'll have to change the code so that the agent is able to use its experience to gradually choose better actions when interacting with the environment!

````python
env_info = env.reset(train_mode=False)[brain_name] # reset the environment
state = env_info.vector_observations[0]            # get the current state
score = 0                                        # initialize the score
i = 0
while True:
    action = np.random.randint(action_size)        # select an action
    env_info = env.step(action)[brain_name]        # send the action to the environment
    next_state = env_info.vector_observations[0]   # get the next state
    reward = env_info.rewards[0]                   # get the reward
    done = env_info.local_done[0]                  # see if episode has finished
    score += reward                                # update the score
    state = next_state                             # roll over the state to next time step
    print("{} - action:{} next_state:{} reward:{} done:{} score:{}".format(i, action, len(next_state), reward, done, score))
    i+=1
    if done:                                       # exit loop if episode finished
        break
    
print("Score: {}".format(score))
````

````python
from collections import deque
import matplotlib.pyplot as plt
import importlib
import torch
import dqn_agent as dq
importlib.reload(dq)
````

Importlib was used to easily reload package so changing neural network architecture in code won't need closing and reloading jupyter notebook. Code below (method learn_dqn) and imported dqn_agent are both modified version of code provided in Value-Based Method part of Deep Reinforcement Learning Nanodegree, Lesson "Deep Q-Network" originally provided for solving OpenAI Gym's LunarLander-v2 environment.

````python
def learn_dqn(env, n_episodes=1000, eps_start=1.0, eps_end=0.05, eps_decay=0.95, scores_window_len = 100):
    scores = []
    mean_scores= []
    scores_window = deque(maxlen=scores_window_len)
    eps = eps_start
    env_info = env.reset(train_mode=True)[brain_name]
    state = env_info.vector_observations[0]
    
    agent = dq.Agent(state_size=state_size, action_size=action_size, seed=0)
    print("agent can do in env with {} states while performin one of {} actions".format(agent.state_size, agent.action_size))
    print(agent.qnetwork_local)
    
    for i_episode in range(1, n_episodes+1):
        env.reset(train_mode=True)[brain_name]
        score = 0
        while True:
            action = agent.act(state, eps)
            action = action.astype(int)
            env_info = env.step(action)[brain_name]
            next_state = env_info.vector_observations[0] 
            reward = env_info.rewards[0]
            done = env_info.local_done[0]
            agent.step(state, action, reward, next_state, done)
            state = next_state
            score += reward
            #print("action:{} next_state:{} reward:{} done:{} score:{}".format(action, len(next_state), reward, done, score))
            if done:
                break
        scores_window.append(score)
        scores.append(score)
        eps = max(eps_end, eps_decay*eps)
        print('\rEpisode {}\tAverage Score: {:.2f}'.format(i_episode, np.mean(scores_window)), end="")
        if i_episode % scores_window_len == 0:
            mean_scores.append(np.mean(scores_window))
            print('\rEpisode {}\tAverage Score: {:.2f}'.format(i_episode, mean_scores[-1]))
        if np.mean(scores_window)>=13.0:
            print('\nEnvironment solved in {:d} episodes!\tAverage Score: {:.2f}'.format(i_episode, np.mean(scores_window)))
            torch.save(agent.qnetwork_local.state_dict(), 'checkpoint.pth')
            break
    return scores, np.mean(scores), mean_scores, agent
````

After short test, it seems that agent is able to learn, now important question is what hyperparameters should be chosen before learning? I think that preferred method should include both hyperparameters optimization method like RandomSearch + some intuition about our banana environment. Due to limited resources let's try grid search with easily accessible parameters such as epsilon decay and minimal epsilon value - same values we will tested 10 times to make it a bit more robust.

````python
final = []

for ee in {0.2, 0.1, 0.05, 0.01}:
    for ed in {0.99, 0.95, 0.90, 0.85, 0.8}:
        ms = {"epsilon_end" : ee, "epsilon_decay" : ed, "scores":[]}
        for i in range(10):
            scores, mean_score, mean_scores, agent = learn_dqn(env, n_episodes=100, eps_end=ee, eps_decay=ed)
            ms["scores"].append(mean_score)
        final.append(ms)
````
